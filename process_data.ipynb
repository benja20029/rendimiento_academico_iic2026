```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('rendimiento_2010-2022.csv', sep=',', encoding='latin-1')
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.dtypes
```

```python
df
```

```python
# Assuming df is your DataFrame
df = df[~df['COD_DEPE'].isin([5, 6])]

# Replace the codes in 'COD_DEPE' with their corresponding institution type
df['COD_DEPE'] = df['COD_DEPE'].replace({1: 'Municipal', 2: 'Municipal', 3: 'Particular Subvencionado', 4: 'Particular Pagado'})
df['COD_ENSE'] = df['COD_ENSE'].replace({110: 'Educacion Basica', 310: 'Educacion Media', 410: "Educacion Media", 510: "Educacion Media", 610: "Educacion Media", 710: "Educacion Media",
                                         810: "Educacion Media", 910: "Educacion Media"})
df['COD_REG_RBD'] = df['COD_REG_RBD'].replace({
    15: "Región de Arica y Parinacota",
    1: "Región de Tarapacá",
    2: "Región de Antofagasta",
    3: "Región de Atacama",
    4: "Región de Coquimbo",
    5: "Región de Valparaíso",
    6: "Región del Libertador Bernardo O'Higgins",
    7: "Región del Maule",
    16: "Región de Ñuble",
    8: "Región del Bío-Bío",
    9: "Región de La Araucanía",
    14: "Región de Los Ríos",
    10: "Región de Los Lagos",
    11: "Región de Aysén del Gral. Ibañez del Campo",
    12: "Región de Magallanes y Antártica Chilena",
    13: "Región Metropolitana de Santiago",
})
# Now group by 'AGNO', 'COD_DEPE', 'COD_REG_RBD', and 'COD_ENSE', and compute the mean of 'PROM_GRAL'
new_df = df.groupby(['AGNO', 'COD_DEPE', 'COD_REG_RBD', 'COD_ENSE'])['PROM_GRAL'].mean().reset_index()
```

```python
new_df
```

```python
## Export to csv
new_df.to_csv('resumen_rendimiento_2010-2022.csv', index=False)
```

```python
## Now we can see only the rows asociated to the 'COD_REG_RBD' == "Region de Atacama"
new_df[new_df['COD_REG_RBD'] == "Atacama"]
```

```python
# Now group by 'AGNO' and 'COD_DEPE', and compute the mean of 'PROM_GRAL'
df_prom_gral_comb = df.groupby(['AGNO', 'COD_DEPE'])['PROM_GRAL'].mean().reset_index()

# Rename the column
df_prom_gral_comb.rename(columns={'PROM_GRAL': 'PROM_GRAL_COMB'}, inplace=True)
```

```python
## Export as csv
df_prom_gral_comb.to_csv('promedio_general_por_tipo_institucion_2010-2022.csv', index=False)
```
